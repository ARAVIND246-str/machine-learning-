```python
# Twitter Sentiment Analysis with Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 1. Load dataset
df = pd.read_csv("twitter_training.csv")

print("Dataset Shape:", df.shape)
print(df.head())

# --- Dataset format: ID,Entity,Sentiment,Tweet ---
X = df.iloc[:, 3]   # Tweet text
y = df.iloc[:, 2]   # Sentiment

# FIX missing values
X = X.fillna("")

# 2. Sentiment distribution visualization
plt.figure(figsize=(6,4))
sns.countplot(x=y, palette="viridis")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# 3. TF-IDF Vectorization
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
X_tfidf = tfidf.fit_transform(X)

# 4. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# 5. Train model (Naive Bayes)
model = MultinomialNB()
model.fit(X_train, y_train)

# 6. Predictions
y_pred = model.predict(X_test)

# 7. Evaluation
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# 8. Confusion Matrix Visualization
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=model.classes_,
            yticklabels=model.classes_)
plt.title("Confusion Matrix Heatmap")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# 9. Custom test
sample_tweet = ["I hate this update, it's the worst ever!"]
sample_vector = tfidf.transform(sample_tweet)
prediction = model.predict(sample_vector)
print("\nCustom Tweet Prediction:", prediction[0])
```

```python
# Robust Email Analysis - auto-detect text/label columns + visualizations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder

# 1) Load dataset
df = pd.read_csv("Email Analysis Dataset.csv")
print("Columns found in the dataset:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())

# 2) Auto-detect text column (heuristic)
text_preferred_names = ['text','message','body','email','content','message_text','message_body','mail','clean_text','subject']
text_cols = [c for c in df.columns if df[c].dtype == 'O']  # object/string columns

X_col = None
for name in text_preferred_names:
    if name in df.columns:
        X_col = name
        break

if X_col is None and text_cols:
    # choose string column with largest average length (likely the email body)
    avg_len = {c: df[c].astype(str).map(len).mean() for c in text_cols}
    X_col = max(avg_len, key=avg_len.get)

if X_col is None:
    raise ValueError("No text-like column found. Please open the CSV and tell me which column contains the email body.")

# 3) Auto-detect label column (heuristic)
label_preferred_names = ['label','spam','target','class','category','type','y','tag','labelled','Category']
y_col = None
for name in label_preferred_names:
    if name in df.columns:
        y_col = name
        break

if y_col is None:
    # pick column (not the text column) with few unique values >1 (likely labels)
    cand = []
    for c in df.columns:
        if c == X_col:
            continue
        nunique = df[c].nunique(dropna=True)
        if 1 < nunique <= 50:  # <=50 so it can still be multi-class but not an id
            cand.append((c, nunique))
    if cand:
        # choose the column with the fewest unique values (likely label)
        y_col = sorted(cand, key=lambda x: x[1])[0][0]
    else:
        # fallback: pick any non-id column (last one)
        possible = [c for c in df.columns if c != X_col and 'id' not in c.lower()]
        if possible:
            y_col = possible[-1]
        else:
            raise ValueError("No label-like column found. Please specify which column is the label.")

print(f"\nDetected text column: '{X_col}'")
print(f"Detected label column: '{y_col}'")

# 4) Prepare data: keep just these two columns and clean
data = df[[X_col, y_col]].copy()
data[X_col] = data[X_col].fillna("").astype(str)
data = data.dropna(subset=[y_col])  # drop rows where label is missing

# 5) Encode labels
le = LabelEncoder()
y = le.fit_transform(data[y_col].astype(str))
label_names = le.classes_
print("\nLabel mapping (encoded -> original):")
for i, name in enumerate(label_names):
    print(i, "->", name)

# 6) TF-IDF vectorization
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
X_tfidf = tfidf.fit_transform(data[X_col])

# 7) Train/test split (stratify if possible)
try:
    X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42, stratify=y)
except Exception:
    X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# 8) Train model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# 9) Evaluation
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {acc:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=label_names))

# 10) Visualizations
plt.figure(figsize=(7,4))
sns.countplot(x=pd.Series(y).map(lambda v: label_names[v]))
plt.title("Label distribution (full dataset)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=label_names, yticklabels=label_names, cmap="Blues")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

# 11) ROC curve if binary
if len(label_names) == 2:
    from sklearn.metrics import roc_curve, auc
    y_score = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_score)
    auc_score = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr)
    plt.plot([0,1],[0,1], '--', linewidth=0.7)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC curve (AUC = {auc_score:.3f})")
    plt.tight_layout()
    plt.show()

# 12) Top features for interpretation
feature_names = tfidf.get_feature_names_out()
if model.coef_.ndim == 1 or model.coef_.shape[0] == 1:
    coef = model.coef_[0]
    top_pos = np.argsort(coef)[-20:][::-1]
    top_neg = np.argsort(coef)[:20]
    print("\nTop features (positive class):", [feature_names[i] for i in top_pos])
    print("Top features (negative class):", [feature_names[i] for i in top_neg])
else:
    for i, cls in enumerate(label_names):
        topn = np.argsort(model.coef_[i])[-20:][::-1]
        print(f"\nTop features for class '{cls}':")
        print([feature_names[j] for j in topn])

# 13) Example prediction
sample_email = ["Congratulations! You have won a free ticket  click to claim now!"]
sample_vec = tfidf.transform(sample_email)
pred_enc = model.predict(sample_vec)
print("\nSample prediction:", le.inverse_transform(pred_enc)[0])
```
